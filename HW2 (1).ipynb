```python
# And then we are going to import libraries that we will use...
import wbgapi as wb
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# We are going to write a function that takes a country 
# name and a list of indicators and returns a pandas data frame with the corresponding data.
def function1 (country , workSpace = {'api1': {'link': 'SI.POV.NAHC'},
             'api2': {'link': 'SH.XPD.CHEX.PC.CD'},
             'api3': {'link': 'SL.TLF.CACT.FE.NE.ZS'},
             'api4': {'link': 'SH.DYN.MORT'},
             'api5': {'link': 'SE.ADT.LITR.FE.ZS'}}):
                country = ['TUR']

                for x,y in workSpace.items():
                    link = y['link']
                    y.update({'data': wb.data.DataFrame(link,economy = country)})
                    
                return wb.data.DataFrame(link,economy = country)
```
